## Solution

The goal of the project is to make the students search and find ways to test Airflow Dags, as it differs from basic testing where we only use native python functions (length of an array, equality of a value...). Here, they will need to learn about how to make Airflow validate the user-defined dags, even without actually loading it for scheduling.

This function or a similar one should exist in the student project :

```python
import json
from airflow.models import DagBag
from os import path

def test_dagbag_import(self):
    dagbag = DagBag(dag_folder=path.join('.', 'dags'), include_examples=False)
    nb_errors = len(self.dagbag.import_errors)
    if len(self.dagbag.import_errors) > 0:
        print(f'{nb_errors} error(s) have been detected!')
        print(json.dumps(self.dagbag.import_errors))

    assert nb_errors == 0
```

This code is actually used in most of production Airflow setups I've seen.
