## Volumes

In this exercise, we will learn how to manipulate volumes.

1. As the excercice 1, launch a postgres database. The postgres database should have:
- persistent data

- the database must be named devops
2. Create a table and add some data

3. Delete the container and recreate it.

4. In your opinion, is the data lost?

For information about the postgres database container, see the dockerhub documentation page.

## Correction

````
%%bash
docker run -d \
    --name postgres \
    -e POSTGRES_PASSWORD=pass \
    -e PGDATA=/var/lib/postgresql/data/pgdata \
    -v pgdata:/var/lib/postgresql/data \
    postgres
````

1. the "data" volume will be created on the fly

2. ```bash
   docker exec -it postgres psql -U postgres;
   postgres=# create table users(name VARCHAR(50));
   postgres=# insert into users values ('jacques');
   postgres=# select * from users;
     name   
   ---------
    jacques
   (1 row)
   ```

3. ```bash
   docker rm -f postgres;
   ```

4. ```bash
   docker run -d \
       --name postgres \
       -e POSTGRES_PASSWORD=pass \
       -e PGDATA=/var/lib/postgresql/data/pgdata \
       -v pgdata:/var/lib/postgresql/data \
       postgres
   ```

            *pgdata* will be reused (already created)

5. ```bash
   docker exec -it postgres psql -U postgres;
   postgres=# \d
   psql (13.0 (Debian 13.0-1.pgdg100+1))
   Type "help" for help.
   
   postgres=# \d
            List of relations
    Schema | Name  | Type  |  Owner   
   --------+-------+-------+----------
    public | users | table | postgres
   (1 row)
   
   postgres=# select * from users;
     name   
   ---------
    jacques
   (1 row)
   ```
