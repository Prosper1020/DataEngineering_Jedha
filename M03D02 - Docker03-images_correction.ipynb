# Image creation

The aim of this exercise is to manipulate, create and modify an image.

1. Launch an alpine container in its latest version with an interactive shell.

2. Install the curl command using apk and test it with `curl www.google.com`

3. Create an image using the instructions you used in the alpine image to install curl.

4. List your images and check that it is present

5. Test the installation of your image in a single command line.

## Correction

1. ```bash
   docker container run -it alpine:latest
   ```

2. ```bash
   apk update; apk add curl; curl www.google.com
   ```

3. ```dockerfile
   FROM alpine:latest
   RUN apk update
   RUN apk add curl
   
   ---
   
   docker build -t alpinecurl:alpha .;
   ```

4. ```bash
   docker image ls;
   ```

5. ```bash
   docker container run alpinecurl:alpha curl www.google.com
   ```
